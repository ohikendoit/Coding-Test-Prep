```python
#Code02-04.py

def multi(v1, v2):
  retList = [] #반환할, 답을 담고있는 리스트 생성
  res1 = v1 + v2
  res2 = v1 - v2
  retList.append(res1)
  retList.append(res2)
  return retList

myList = []
hap, sub = 0, 0

myList = multi(100, 200)
hap = myList[0]
sub = myList[1]
print("multi()에서 반환한 값 ==> %d, %d" % (hap, sub))
```
